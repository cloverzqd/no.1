# 下程序为封装两个函数进行两次清洗
（注意第一次清洗时需要以str形式判断非数值，所以打开read_csv时候要用转成str形式）
（而二次清洗时，需要进行数值比较，在函数内加上.astype(int)再变回数值形式）
此程序用seaborn画图，彩色的

此程序与下一个程序的图有区别的原因是：此程序是先再打开csv时已经cut_list选取了部分列之后再筛选空值
所以筛去后所剩的数据比较多，

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
#________________函数_________________________
def clean_null(df):
    print('筛去空值ing...')
    col_list =df.columns
    #df1=df[df.notnull()]#保留有数据的行
    for i in col_list:
        df_bool=df[i].str.isdigit() 
    #i可直接用 ['ctrl_maxel_max']列名代替，即只筛选第一列 即可把改文件中一整行都错的数据删掉
    #循环是把只要出现错误的行都删掉
    df_cleaned = df[df_bool].copy() 
    #print(df_cleaned)
    return df_cleaned

def cleaning(df,list):  
    print('过滤上下限...')
    G = globals()
    for x in range(len(list)):
    #for x in range(0,3):
        # list[x][0] :属性名称
        # list[x][1] :上限
        # list[x][2] :下限
        df = df[(df[list[x][0]].astype(int)  >= list[x][1]) &
                (df[list[x][0]].astype(int)  <= list[x][2]) ]
    return df

#________________主程序_________________________
cut_list = [ # 一个Python的列表，用于记录需要读取的属性
    'gbox_hsst_avg',
    'grid_ap_avg',
    'cvet_grs_avg', # 转速
    'wind_spd_avg', # 有功
    'evmt_cbot_avg',
    ]
#df = pd.read_csv(file_path, index_col='rtime', usecols = col_list , parse_dates=True)
#df
path = r'E:\EPRI\BeiChe_Tielin\风机No.31\wtg_31_1.csv'  
f= open(path,encoding="utf-8")  
df= pd.read_csv(f, usecols = cut_list,dtype=str)#,dtype=str
#df1=df[df.notnull()]#保留有数据的行
#print(df)
#----------一次清洗空值--------
df_clean1=clean_null(df)
#print(df_clean1)
print('after cleanning data length =',len(df_clean1))
#----------二次清洗范围内值--------
cleaning_list = [
    ['wind_spd_avg',0,300],
    ['cvet_grs_avg',0,20000],
    ['evmt_cbot_avg',-200,1500],
    ['grid_ap_avg',-200,30000]
    # 按照以上格式添加规则
    # ['名称',最小值,最大值],
    ]

df_clean2 =cleaning(df_clean1,cleaning_list)

print(df_clean2)
print('after cleanning data length =',len(df_clean2))

y1=df_clean1['grid_ap_avg'].astype(int) #纵轴功率
x1=df_clean1['wind_spd_avg'].astype(int)#横轴
y2=df_clean2['grid_ap_avg'].astype(int) #纵轴功率
x2=df_clean2['wind_spd_avg'].astype(int)#横轴
plt.figure(figsize=(20,20))
plt.subplot(211)
s1 = sns.stripplot(x1,y1) #, color='r', s=25

plt.subplot(212)
s2 = sns.stripplot(x2,y2)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
#________________定义函数_________________________
def cleaning(df,list):  
    print('过滤上下限...')
    G = globals()
    for x in range(len(list)):
    #for x in range(0,3):
        # list[x][0] :属性名称
        # list[x][1] :上限
        # list[x][2] :下限
        df = df[(df[list[x][0]].astype(int) >= list[x][1]) &
                (df[list[x][0]].astype(int) <= list[x][2]) ]
    return df


#________________主程序_________________________
#筛选空值的列的列表：


path = r'E:\EPRI\BeiChe_Tielin\风机No.31\wtg_31_1.csv'  
f= open(path,encoding="utf-8")  
df= pd.read_csv(f,dtype=str)#

col_list =df.columns
#------一次清洗，筛去空值-------
df1=df[df.notnull()]#保留有数据的行
for i in col_list:
    df_bool=df1[i].str.isdigit() 
    #i可直接用 ['ctrl_maxel_max']列名代替，即只筛选第一列 即可把改文件中一整行都错的数据删掉
    #循环是把只要出现错误的行都删掉
df_clean1 = df1[df_bool].copy() 
print("筛去空值ing...")
#print(df_clean1)
print('筛去空值后data length =',len(df_clean1))


#-------二次清洗，范围内值--------
cut_list = [ # 一个Python的列表，用于记录需要读取的属性
    'gbox_hsst_avg',
    'grid_ap_avg',
    'cvet_grs_avg', # 转速
    'wind_spd_avg', # 有功
    'evmt_cbot_avg',
    ]

cleaning_list = [
    ['wind_spd_avg',0,300],
    ['cvet_grs_avg',0,20000],
    ['evmt_cbot_avg',-200,1500],
    ['grid_ap_avg',-200,30000]
    # 按照以上格式添加规则
    # ['名称',最小值,最大值],
    ]

df_clean2 =cleaning(df_clean1,cleaning_list)
#print(df_clean2)
print('过滤上下限后 data length =',len(df_clean2))

y1=df_clean1['grid_ap_avg'].astype(int) #纵轴功率
x1=df_clean1['wind_spd_avg'].astype(int)#横轴.astype(int)
y2=df_clean2['grid_ap_avg'].astype(int) #纵轴功率
x2=df_clean2['wind_spd_avg'].astype(int)#横轴风速
#-----------绘图------------
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号  

plt.figure(figsize=(20,20))#figsize=(80,60)figsize=(30,20)
axe1 = plt.subplot(211)
s1 = axe1.scatter(x1,y1, color='r', s=25)
plt.title(u'清洗前功率散点图')
plt.xlabel(u'风速wind_spd_avg')
plt.ylabel(u'功率grid_ap_avg')
plt.legend([s1],[r'清洗前'])

axe2 = plt.subplot(212)
s2 = axe2.scatter(x2,y2, color='b', s=25)
plt.title(u'清洗后功率散点图')
plt.legend([s2],[r'清洗后'])
plt.xlabel(u'风速wind_spd_avg')         
plt.ylabel(u'功率grid_ap_avg')

plt.show()
```
