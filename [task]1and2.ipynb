# 2.5 绘制单台风机写成函数，用for循环调用：

```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt 
import pylab as pl

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签 
#定义函数
def draw_pic(x):
    path='E:\EPRI\BeiChe_Tielin\GenData\wtg_%s.csv' %(x)
    print(path)
    f= open(path,encoding="utf-8")  
    df = pd.read_csv(f, index_col='rtime', parse_dates=True)    
    #df=df[1000:len(df)]#对付空值
    shuxing='gbox_lsst_avg'
    y=df[shuxing] 
    dataframe = pd.DataFrame(y)
    df['gbox_lsst_avg'].astype(int)
    #df.columns   #不知道有哪些列，打一下

    pic=dataframe.plot(figsize=(8,6),title='风机：%s  属性：%s'%(x,shuxing))
    plt.xticks(rotation=45)#x刻度数值旋转90° 
    plt.show()
    return pic;

wt_num_list=[32,33,68]#列表循环一些风机号
for i in wt_num_list:
    draw_pic(i)
```
